## **Import libary**

```
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score

from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.models import Sequential

from tensorflow.python.keras.layers import Flatten
from tensorflow.python.keras.layers import Dropout

import random
import numpy as np

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score,r2_score,mean_absolute_error,mean_squared_error
```

## **Load data and extract Global_active_power**

```
file='/content/household_power_consumption.txt'
import pandas as pd
data = pd.read_csv(file, sep=';')
data_=data["Global_active_power"].values
data=[]
count=0
for i in data_:
  if i=="?":
    i=np.nan
    count+=1
  data.append(i)
data=np.array(data)
rg=len(data)-61
print(count)
```

## **Take 60000 random sample from data to train model**

```
X_train=[]
Y_train=[]
count=0
limit = 60000
while count < limit:
  i= np.random.randint(low=0, high=rg)
  temp=data[i:i+61]
  temp = np.float32(temp)
  if(np.isnan(temp).any()):
    continue
  X_train.append(temp[:-1])
  Y_train.append(temp[60])
  count+=1
X_train=np.array(X_train)
Y_train=np.array(Y_train)
Y_train= np.float64(Y_train)
```

## **Fit model with 2 hidden layer(100,80)**

```
model = Sequential()
model.add(Dense(100, activation ='relu', input_dim = X_train.shape[1]))
model.add(Dense(80, activation = 'relu'))
model.add(Dense(1,activation = 'relu'))

model.compile(loss="mse", optimizer='adam', metrics=['mse','mae'])

model.fit(X_train,Y_train, epochs=100, batch_size=100,verbos=1)
```

## **Take random 10000 sample from data to test**

```
X_test=[]
Y_test=[]
count=0
limit = 10000
while count < limit:
  i= np.random.randint(low=0, high=rg)
  temp=data[i:i+61]
  temp = np.float64(temp)
  if(np.isnan(temp).any()):
    continue
  X_test.append(temp[:-1])
  Y_test.append(temp[60])
  count+=1
X_test=np.float64(X_test)
Y_test= np.float64(Y_test)
X_test=np.array(X_test)
Y_test=np.array(Y_test)
Y_test=Y_test.reshape([10000,1])
```

## **R2 score and MSE**

```
from keras import losses
ynew = model.predict(X_test)
from sklearn.metrics import r2_score
print("R2 score = ",r2_score(Y_test,ynew))
print("Mean Absolute Error :",mean_absolute_error(Y_test,ynew))
print("Mean Square Error :",mean_squared_error(Y_test,ynew))
```
